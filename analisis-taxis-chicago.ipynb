Para esta hip√≥tesis, vamos a definir lo siguiente:

Hip√≥tesis Alternativa (H1): La duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare es diferente un s√°bado lluvioso a un s√°bado de buen clima.

```python
#Importar librer√≠a necesaria 
from scipy import stats as st

#Cargar el dataset 
trips_weather = pd.read_csv('/datasets/project_sql_result_07.csv')

#Visualizar dataset
print(trips_weather.info())
print()
print(trips_weather.head())
```

```python
#Separar los datos en buenas y malas condiciones clim√°ticas 
good_weather_condition = trips_weather[trips_weather['weather_conditions']=='Good']['duration_seconds']
bad_weather_condition = trips_weather[trips_weather['weather_conditions']=='Bad']['duration_seconds']
```

<div class="alert alert-block alert-warning"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Al realizar una prueba de hip√≥tesis, es muy importante verificar los supuestos. El t-test que usaste funciona bien cuando se cumplen ciertos supuestos, como que las varianzas de ambos grupos (s√°bados lluviosos y no lluviosos) sean similares. Puedes realizar la prueba de levene para comprobar que los dos grupos tienen la misma varianza y, de esta forma, podr√°s saber el valor del par√°metro equal_var que se debe utilizar en la funci√≥n stats.ttest_ind.

</div>

```python
# Definir el nivel de significancia 
alpha = 0.05

# Realizar prueba T Student 
# Se escoge esta prueba porque estamos comparando dos medias de dos poblaciones diferentes 
results = st.ttest_ind(good_weather_condition, bad_weather_condition)

print("p_value: ", results.pvalue)


if results.pvalue < alpha:
    print("Rechazamos la hip√≥tesis nula: La duraci√≥n promedio de viaje cambia significativamente dependiendo del clima")
else:
    print("No es posible rechazar la hip√≥tesis nula: No hay evidencia suficiente de que el clima afecte la duraci√≥n de un viaje")
```

</div>

- Se usa el ttest_ind porque es la herramienta para comparar dos grupos y saber si son estad√≠sticamente diferentes (viajes con sol vs viajes con lluvia)
- Dado que el valor p es menor que el valor de significancia establecido en 0.05, se rechaza la hip√≥tesis nula, lo que permite afirmar que existe una diferencia significativa en la duraci√≥n de un viaje de un s√°bado lluvioso versus la duraci√≥n de un viaje de un s√°bado soleado o de buen clima. Este resultado era de esperarse, ya que en la mayor√≠a de ciudades, las condiciones clim√°ticas aumentan la duraci√≥n de los viajes, ya que suele haber m√°s tr√°fico, quizas debido a las precauciones extras que se toman en una tormenta o lluvias.
